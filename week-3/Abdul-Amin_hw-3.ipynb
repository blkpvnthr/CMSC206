```python
import math

# Author: Asmaa Abdul-Amin

# Function to calculate area of different shapes
def calculate_area():
    shape = input("Enter the shape (circle, rectangle, triangle): ").strip().lower()

    # Check if the shape is a circle
    if shape == "circle":
        try:
            # Prompt user to enter the radius
            radius = float(input("Enter the radius: "))

            # Validate that the radius is a positive number
            if radius <= 0:
                print("Invalid input: Radius must be a positive number.")
                return
            
            # Calculate and print the area of the circle using the formula πr²
            print(f"The area of the circle is: {round(math.pi * radius ** 2, 2)}")

        except ValueError:
            # Handle invalid input if the user enters non-numeric data
            print("Invalid input: Please enter a numerical value for the radius.")

    # Check if the shape is a rectangle
    elif shape == "rectangle":
        try:
            # Prompt user to enter length and width
            length = float(input("Enter the length: "))
            width = float(input("Enter the width: "))

            # Validate that both length and width are positive numbers
            if length <= 0 or width <= 0:
                print("Invalid input: Length and width must be positive numbers.")
                return
            
            # Calculate and print the area of the rectangle using the formula length × width
            print(f"The area of the rectangle is: {round(length * width, 2)}")

        except ValueError:
            # Handle invalid input if the user enters non-numeric data
            print("Invalid input: Please enter numerical values for length and width.")

    # Check if the shape is a triangle
    elif shape == "triangle":
        try:
            # Prompt user to enter base and height
            base = float(input("Enter the base: "))
            height = float(input("Enter the height: "))

            # Validate that both base and height are positive numbers
            if base <= 0 or height <= 0:
                print("Invalid input: Base and height must be positive numbers.")
                return
            
            # Calculate and print the area of the triangle using the formula (1/2) × base × height
            print(f"The area of the triangle is: {round(0.5 * base * height, 2)}")

        except ValueError:
            # Handle invalid input if the user enters non-numeric data
            print("Invalid input: Please enter numerical values for base and height.")

    # Handle case where user enters an invalid shape
    else:
        print("Invalid shape. Please choose from circle, rectangle, or triangle.")


# Function to compute sum, average, max, and min of an unlimited number of arguments
def foo():
    try:
        # Replace commas with spaces to allow both formats
        user_input = input("Enter numbers separated by spaces or commas: ").replace(",", " ")
        
        # Split by spaces and convert to float
        numbers = list(map(float, user_input.split()))
        
        if not numbers:
            print("Invalid input: Please provide at least one number.")
            return

        total_sum = sum(numbers)
        avg = round(total_sum / len(numbers), 2)
        max_value = max(numbers)
        min_value = min(numbers)

        print(f"Sum: {total_sum}, Average: {avg}, Max: {max_value}, Min: {min_value}")

    except ValueError:
        print("Invalid input: Please enter valid numbers separated by spaces or commas.")

# Menu system to let the user choose which function to execute
def main():
    while True:
        print("\n--- Menu ---")
        print("1. Calculate area of a shape")
        print("2. Compute sum, average, max, and min of numbers")
        print("3. Exit")

        choice = input("Select an option (1-3): ").strip()

        if choice == '1':
            calculate_area()
        elif choice == '2':
            foo()
        elif choice == '3':
            print("Exiting the program. Goodbye!")
            break
        else:
            print("Invalid choice. Please enter 1, 2, or 3.")

# Run the main function for user interaction (Not executable in this environment)
if __name__ == "__main__":
    main()

```

```python
import math
import pandas as pd

# Author: Asmaa Abdul-Amin

# Function to calculate area of different shapes
def calculate_area(shape, **kwargs):
    shape = shape.lower()
    
    if shape == "circle":
        radius = kwargs.get("radius")
        if radius is None or radius <= 0:
            return "Invalid input: Radius must be a positive number."
        return round(math.pi * radius ** 2, 2)
    
    elif shape == "rectangle":
        length = kwargs.get("length")
        width = kwargs.get("width")
        if length is None or width is None or length <= 0 or width <= 0:
            return "Invalid input: Length and width must be positive numbers."
        return round(length * width, 2)

    elif shape == "triangle":
        base = kwargs.get("base")
        height = kwargs.get("height")
        if base is None or height is None or base <= 0 or height <= 0:
            return "Invalid input: Base and height must be positive numbers."
        return round(0.5 * base * height, 2)
    
    else:
        return "Invalid shape. Please choose from circle, rectangle, or triangle."

# Function to compute sum, average, max, and min of an unlimited number of arguments
def foo(*args):
    if not args:
        return "Invalid input: Please provide at least one number."
    
    total_sum = sum(args)
    avg = round(total_sum / len(args), 2)
    max_value = max(args)
    min_value = min(args)
    
    return {
        "Sum": total_sum,
        "Average": avg,
        "Max": max_value,
        "Min": min_value
    }

# Test Cases for Area Calculation
area_test_cases = [
    {"shape": "circle", "radius": 5},
    {"shape": "rectangle", "length": 10, "width": 4},
    {"shape": "triangle", "base": 6, "height": 8},
    {"shape": "circle", "radius": -3},  # Invalid case
    {"shape": "rectangle", "length": 0, "width": 5},  # Invalid case
]

area_results = {
    f"{case['shape'].capitalize()} with {case}": calculate_area(case.pop('shape'), **case)
    for case in area_test_cases
}

# Test Cases for Foo Function
foo_test_cases = [
    (1, 3, 4),
    (5, 10, 15, 20, 25),
    (-2, -4, -6, -8),
    (100, 200, 300),
    ()
]

foo_results = {f"Input: {case}": foo(*case) for case in foo_test_cases}

# Display test results
name="Area Calculation Results" 
df=pd.DataFrame(area_results.items(), columns=["Test Case", "Calculated Area"])
df

```
