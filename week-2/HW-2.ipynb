```python
# Function to prompt for valid input
def get_valid_score():
    while True:
        try:
            score = float(input("Enter a score between 0.0 and 1.0: "))
            if 0.0 <= score <= 1.0:
                return score
            else:
                print("Error: The value must be between 0.0 and 1.0. Please try again.")
        except ValueError:
            print("Error: Invalid input. Please enter a number.")

# Function to assign letter grade based on score
def assign_grade(score):
    if score >= 0.9:
        return 'A'
    elif score >= 0.8:
        return 'B'
    elif score >= 0.7:
        return 'C'
    elif score >= 0.6:
        return 'D'
    else:
        return 'F'

# Get valid score and print corresponding grade
score = get_valid_score()
grade = assign_grade(score)
print(f"Your grade is: {grade}")
```

## 3. Additional Practice Code

```python
# Author: Asmaa Abdul-Amin

# Function to calculate factorial
def factorial(n):
    result = 1
    for i in range(2, n + 1):
        result *= i
    return result

# Prompt user for a number
num = int(input("Enter a non-negative integer to calculate its factorial: "))

if num < 0:
    print("Error: Factorial is not defined for negative numbers.")
else:
    print(f"The factorial of {num} is {factorial(num)}.")
```
